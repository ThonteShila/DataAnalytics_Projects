```python
import nltk
```

```python
nltk.download("vader_lexicon")
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sia=SentimentIntensityAnalyzer()
```

```python
sia
```

```python
comments["comment_text"]
```

```python
sia.polarity_scores("MY FAN . attendance")['compound']
```

```python
sentiments=[]
for i in comments["comment_text"]:
    score=sia.polarity_scores(str(i))['compound']
    sentiments.append(score)
sentiments[0:100]
```

```python
comments["Polarity"]=sentiments
```

```python
comments
```

###wordcloud

wordcloud-
it create img by using +ve sentiments using maptlotlib

```python
filter_pos=(comments["Polarity"]>=0.8) & (comments["Polarity"]<=1.0)
pos_comments=comments[filter_pos]
pos_comments=pos_comments["comment_text"]
pos_comments
```

```python
filter_neg=(comments["Polarity"]>=-1.0) & (comments["Polarity"]<=0.8)
neg_comments=comments[filter_neg]
neg_comments=neg_comments["comment_text"]
neg_comments[0:1000]
```
